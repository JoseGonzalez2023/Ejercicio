```python
## Vamos a crear una función para identificar datos nulos ##
## Las funciones en Python nos permiten crear un bloque se cógico utilizable que hace una tarea específica, los que nos permite llamar a auna función 
## las veces que las necesitemos solo con llamarla, las funciones en Python las creamos con la función def, luego a continuación se debe escribir el 
## nombre de la variable y nunca debe comenzar con números, y luego entre parentesis se escribe los paremtros del codigo que se necesita## 


## Ejemplo, vamos a crear una función que sume 2 numeros ##
def sumar_dos_numeros(x,y):
  resultado=x+y
  return resultado

```

```python
def identificar_datos_nulos(df):
  variables=df.columns.tolist()
  nulos=[] 
  for i in variables:  
    nulos.append(df[i].isnull().sum())
  tabla=pd.DataFrame({"variables":variables,"valores nulos":nulos})
  tabla["porcentaje de nulos"]=tabla["valores nulos"]/len(df)*100
  return tabla.sort_values("valores nulos",ascending=False)
  
```

```python
nulos=[] ## Vamos a guardar en esta lista vacía los valores nulos de cada variable, cada vez que el FOR está interando por cada una de las variables, se elimana Print
for i in variables : ##Acceder a la lista con un for##
  print(df[i].isnull().sum())
  nulos.append(df[i].isnull().sum())  ## se va guardando la informaciómn en la lista de nulos cada vez que se itera
```

```python
nulos ##Se muestra los valores nulos##
```

```python
tabla=pd.DataFrame({"variables":variables,"valores nulos":nulos}) ##creamos una tabla con pandas pd, utizamos DataFrame y entre parentesis se agregan el dicccionario con llaves de las columnas que queremos agregar.
```

```python
tabla=tabla.sort_values("valores nulos",ascending=False) ##La tabla muestra ordenado de mayor a menor las variables de valores nulos
```

```python
tabla
```

```python
##vamos a crear el vector de nulos##

tabla["porcentaje de nulos"]=tabla["valores nulos"]/len(df)*100
tabla
```

```python
##identificamos las variables que tienen más de 60% que tengan variables nulos##
nulas_a_eliminar=tabla[tabla["porcentaje de nulos"]>=60]["variables"].to_list()
nulas_a_eliminar
```
